```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
columns_name = ['매물명', '동', '거래유형', '가격', '유형', '전체면적', '실제면적',
                    '높이', '방향', '설명', '특징', '확인', '확인날짜', '링크', '크롤링날짜']
data = pd.read_csv("../부동산/화명동(2022-06-27(12,34,44)).csv", names=columns_name)
data.head(5)
```

```python
def number(a):
    if '/' in a:
        a = np.nan
        return a
    a = a.replace('~',' ')
    a = a.replace(',','')
    num = a.split()
#     print(len(num)) 
#     print(num)
    
    if len(num) ==  1: 
        num = num[0]
        if '억' in num: # 억이 num[0]에 들어가면 억을 00000000로 바꿈
            num = int(num.replace('억','00000000'))
            return num 
        else :
            return int(num + '0000') #억이 없는 num[0]은 int로 바꾸면서 뒤에 0000 추가함
        
        
    elif len(num) == 2:
        if '억' not in num[0]: # 억이 포합되지 않으면 num[0]앞에 값을 int로 변환후 0000을 추가해서 출력
            return int(num[0] + '0000')
    
        if '억' in num[0]:
            num[0] = int(num[0].replace('억', '00000000')) 
            if '억'in num[1]: # 억이 num[1]에 들어가면 num1에 있는 억을 00000000로 바꿔서 num[1]값을 출력함
                return num[1].replace('억','00000000')
            num[1] = int(num[1] + '0000')
            num = int(num[0] + num[1])
        return num
    
    
    
    elif len(num) == 3:
        del num[0] 
        if '억' in num[0]:
            num[0] = int(num[0].replace('억','00000000'))
            num[1] = int(num[1] + '0000')
            num = num[0] + num[1]
            return num
                  

    

             
    elif len(num) == 4:
        if '억' in num[0]:
            num[0] = int(num[0].replace('억', '00000000')) 
            num[1] = int(num[1]+'0000')
            num1 = num[0] + num[1] # num1에 num[0]이랑 num[1]을 정수로 바꿔서 합한 값을 저장
        if '억'in num[2]:
            num[2] = int(num[2].replace('억', '00000000'))
            num[3] = int(num[3]+'0000')
            num2 = num[2] + num[3] # num2에 num[2]이랑 num[3]을 정수로 바꿔서 합한 값을 저장
        numhap = (num1+num2)/2 # num1과 num2를 합해서 2로 나눠 중간값을 만듬
        return int(numhap)    

    else:
        return num
```

```python
data['전처리한아파트가격'] = data['가격'].apply(number)
```

```python
def date(a): # 22가 들어가있는것들은 2022로 변경
    a = a.split('.')
    print(type(a))
    if '22' in a[0]:
        a[0] = a[0].replace('22','2022')
        a = '.'.join(a)
        return a
    else: 
        return '.'.join()
```

```python
data['확인날짜'] = data['확인날짜'].apply(date)
```
